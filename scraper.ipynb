## Username

```python
username = str(input("Enter the Player's username: ")).strip()
link = "https://www.chess.com/stats/overview/"+username

#print(link)
```

## Web Driver Setup

```python
service =Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)

driver.get(link)
#driver.get(wrong_link)
```

```python
soup =BeautifulSoup(driver.page_source,'html.parser')
soup
```

## Verifying Player  Username

```python
sp = soup.find('div',class_="_404-header")
temp = sp.find('h1').text
if temp=="404 Page not found":
    print("Invalid Username")
```
